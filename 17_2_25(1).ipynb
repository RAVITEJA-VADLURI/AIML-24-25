```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
N = 500
children_counts = np.random.choice([0, 1, 2, 3, 4], size = N, p = [0.1, 0.25, 0.35, 0.2, 0.1])
```

```python
unique, counts = np.unique(children_counts, return_counts = True)
pmf = counts/N
```

```python
cdf = np.cumsum(pmf)
plt.figure(figsize=(10, 5))
sns.barplot(x = unique, y = pmf, color = 'blue', edgecolor = 'white')
plt.xlabel('Number of Children')
plt.ylabel('Probability (PMF)')
plt.title('Probability Mass Function (PMF) of Family Sizes')
plt.xticks(unique)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(x = unique, y = cdf, marker = 'o', color = 'red', linestyle = '-')
plt.xlabel('Number of Children')
plt.ylabel('Cumulative Probability (CDF)')
plt.title('Cumulative Distribution Function (CDF) of Family Sizes')
plt.xticks(unique)
plt.ylim(0, 1.1)
plt.show()
```

```python
p = 0.20
N = 100
prob_25 = stats.binom.pmf(25, N, p)
prob_35 = 1 - stats.binom.cdf(34, N, p)
expected_clicks = N*p
print(f"Probability of 25 clicks: {prob_25:.4f}")
print(f"Probability of more than 35 clicks: {prob_35:.4f}")
print(f"Expected number of clicks: {expected_clicks:.2f}")
```
